### Função para plotar o gráfico com ruído


```python
def lenz_feliz(df, nome): #Deixa o Lenz muito feliz

  ch_0= ch_gen_final = alt.Chart(df, title= f"Temperatura ({nome})").mark_line(strokeDash=[5, 5]).encode(
    alt.X('Tempo (s)', title='Tempo (s)'),
    alt.Y('Temp Desejada (C)',
           scale=alt.Scale(domain=[min(df['Temp Atual (C)']-1),
                                   max(max(df['Temp Atual (C)']),max(df['Temp Desejada (C)'])+1)]),
          title='Temperatura (C)')
  )

  ch_1= ch_gen_final = alt.Chart(df, title= f"Temperatura ({nome})").mark_line(color='red').encode(
    alt.X('Tempo (s)', title='Tempo (s)'),
    alt.Y('Temp Atual (C)',scale=alt.Scale(domain=[min(df['Temp Atual (C)']-1),
                                                   max(max(df['Temp Atual (C)']),max(df['Temp Desejada (C)']))+1]),title='Temperatura (C)'),
    alt.OpacityValue(0.5)
  )

  ch_2= ch_gen_final = alt.Chart(df, title=f'PWM ({nome})').mark_line().encode(
    alt.X('Tempo (s)', title='Tempo (s)'),
    alt.Y('PWM',title=''),
    alt.OpacityValue(1)
  )

  chs_1= ch_gen_final = alt.Chart(df, title= f"Temperatura ({nome})").mark_line(color='red').encode(
    alt.X('Tempo (s)', title='Tempo (s)', bin=True),
    alt.Y('average(Temp Atual (C))',scale=alt.Scale(domain=[min(df['Temp Atual (C)']-1),
                                                   max(max(df['Temp Atual (C)']),max(df['Temp Desejada (C)']))+1]),title='Temperatura (C)'),
    alt.Tooltip(['average(Temp Atual (C))','average(PWM)'])
  )

  chs_2= ch_gen_final = alt.Chart(df, title=f'PWM ({nome})').mark_line().encode(
    alt.X('Tempo (s)', title='Tempo (s)',bin=True),
    alt.Y('average(PWM)',title=''),
    alt.Tooltip(['average(Temp Atual (C))','average(PWM)'])
  )
#alt.BinParams(maxbins=100)

  return ((ch_1).properties(width=1000, height=400) | (ch_2).properties(width=1000, height=400)).transform_filter(alt.datum["Tempo (s)"] <= 3000)
```
